```python
from experiments import run_experiment_series
from matplotlib import pyplot as plt
from markets import SimplifiedDoubleAuctionMarket

plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 300
```

# Setting the Parameters

```python

# List of numbers of participants/trading rounds
np_list = [3, 5, 10, 50, 100, 500, 1000, 2000, 5000]

# Number of runs of the simulation for each 'number of participants' in the np_list
n_E = 10
```

# Index Market Experiment

```python
config_index = {
    # The kind of market. Is it a winner take all market answering a
    # binary question ('binary') or a regression market extimateing an index
    # ('index')
    'mode': 'index',
    'market_class': SimplifiedDoubleAuctionMarket,
    # The number of experiments to be run for each number of participants/trading rounds
}

xindex, yindex, _ = run_experiment_series(np_list, config_index, n_E)



plt.xscale('log')

plt.plot(xindex, yindex)
plt.xlabel('Distance from true value')
plt.ylabel('Number of participants and trading rounds')
plt.savefig('indexmarket.png')
```

## Collecting the Results into a single figure

```python
fig, ((ax1, ax2), (ax3, ax4), (ax5, ax6)) = plt.subplots(3, 2)
  

fig.set_figheight(15)
fig.set_figwidth(12)
fig.tight_layout(pad=5.0)



fig.supylabel('Probability of reaching the truth')
fig.supxlabel('Number of participants and trading rounds')

for ax in [ax1, ax2, ax3, ax4, ax5, ax6]:
    ax.set_xscale('log')
    ax.set_ylim([.5, 1.02])
for ax in [ax1, ax2]:
    ax.set_ylim([-0.02, 1.02])


# For Sine Function
ax1.plot(x03, y03)
ax1.set_title("pright = 0.3")
  
# For Cosine Function
ax2.plot(x04, y04)
ax2.set_title("pright = 0.4")
  
# For Tangent Function
ax3.plot(x05, y05)
ax3.set_title("pright = 0.5")
  
# For Tanh Function
ax4.plot(x06, y06)
ax4.set_title("pright = 0.6")
  

ax5.plot(x07, y07)
ax5.set_title("pright = 0.7")  

ax6.plot(x08, y08)
ax6.set_title("pright = 0.8")  
plt.savefig('binarymarket.png')
plt.show()
```
